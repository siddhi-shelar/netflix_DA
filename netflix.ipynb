# Dropping the columns

```python
cols=['Overview','Original_Language','Poster_Url']
```

```python
df.drop(cols,axis=1,inplace=True)
df.columns
```

```python
df.head()
```

```python
def catigorize_col(df, col, labels):
  edges = [
      df[col].describe()['min'],
      df[col].describe()['25%'],
      df[col].describe()['50%'],
      df[col].describe()['75%'],
      df[col].describe()['max']
  ]
  # Remove the 'duplicate' argument or update pandas
  df[col] = pd.cut(df[col], edges, labels=labels, duplicates='drop')
  return df
```

```python
labels=['not_popular','below_avg','average','popular']
catigorize_col(df,'Vote_Average',labels)
df['Vote_Average'].unique()
```

##which year has the most filmed movies?

```python
# prompt: which year has the most filmed movies?

# Group the DataFrame by release year and count the number of movies in each year
movies_per_year = df.groupby('Release_Date')['Title'].count()

# Find the year with the most movies
year_with_most_movies = movies_per_year.idxmax()

# Print the result
print(f"The year with the most filmed movies is: {year_with_most_movies}")
```
